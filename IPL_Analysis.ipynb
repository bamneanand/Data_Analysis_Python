# Cleaning Data

```python
data.columns
```

```python
batters = data.loc[(data['Player_Type']=='Batter')]
batters_new = batters.loc[(data['Capped']== 1)]
capped_batters = batters_new[['Player Name', 'Team', 'Nationality',
       'Matches_Played', 'Runs', 'Average', 'Strike_Rate']]
```

```python
capped_batters.head()
```

```python
bowlers = data.loc[(data['Player_Type']=='Bowler ')]
bowlers_new = bowlers.loc[(data['Capped']== 1)]
capped_bowlers = bowlers_new[['Player Name', 'Team', 'Nationality',
       'Matches_Played','Wickets','Bowling_average', 'Economy', 'Bowling_Strike_Rate']]
```

```python
capped_bowlers.head()
```

```python
keepers = data.loc[(data['Player_Type']=='Keeper')]
keepers_new = keepers.loc[(data['Capped']== 1)]
capped_keepers = keepers_new[['Player Name', 'Team', 'Nationality',
       'Matches_Played','Runs', 'Average', 'Strike_Rate','Catches','Run_outs', 'Stumps']]
```

```python
capped_keepers.head()
```

```python
Allroundres = data.loc[(data['Player_Type']=='Allrounder')]
Allroundres_new = Allroundres.loc[(data['Capped']== 1)]
capped_Allroundres = Allroundres_new[['Player Name', 'Team', 'Nationality',
       'Matches_Played','Runs', 'Average', 'Strike_Rate','Wickets','Bowling_average', 'Economy', 'Bowling_Strike_Rate']]
```

```python
capped_Allroundres.head()
```

```python
capped_batters=capped_batters.fillna(0)
capped_bowlers=capped_bowlers.fillna(0)
capped_keepers=capped_keepers.fillna(0)
capped_Allroundres=capped_Allroundres.fillna(0)
```

```python
print(capped_batters.isna().sum())
print(capped_bowlers.isna().sum())
print(capped_keepers.isna().sum())
print(capped_Allroundres.isna().sum())
```
