## Data Analysis
##### 1.A) Analyze the performance of sales and revenue at the city and branch level ( 5 marks)

```python
sales_revenue_city_branch_total = df.groupby(['City', 'Branch'])[['Revenue', 'Sales']].sum().reset_index()
sales_revenue_city_branch_total 
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(sales_revenue_city_branch_total['City'], sales_revenue_city_branch_total['Revenue'], label='Revenue', alpha=0.7)
ax.bar(sales_revenue_city_branch_total['City'], sales_revenue_city_branch_total['Sales'], label='Sales', alpha=0.7)
ax.set_xlabel('City/Branch')
ax.set_ylabel('Amount')
ax.set_title('Sales and Revenue by City/Branch')
ax.legend()
plt.xticks(rotation=90, ha='right')
plt.show()
```

##### 1.B) What is the average price of an item sold at each branch of the city (10 marks)

```python
average_price_per_branch =round(df.groupby(['City', 'Branch'])['Unit price'].mean(),2).reset_index()
average_price_per_branch
```

```python
fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(average_price_per_branch['Branch'], average_price_per_branch['Unit price'], color='skyblue', alpha=0.7)
ax.set_xlabel('Branch')
ax.set_ylabel('Average Unit Price')
ax.set_title('Average Unit Price per Branch')
plt.show()
```

##### 1.C) Analyze the performance of sales and revenue, Month over Month across the Product line, Gender, and Payment Method, and identify the focus areas to get better sales for April 2019. (15 marks)

```python
monthly_sales = df.groupby(['Year', 'Month', 'Product line', 'Gender', 'Payment'])['Sales'].sum().reset_index()
monthly_sales
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Month', y='Sales', hue='Payment', data=monthly_sales, ci=None)
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales by Payment Method')
plt.show()
```
